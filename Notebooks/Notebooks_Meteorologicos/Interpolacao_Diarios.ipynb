```python
import matplotlib.pyplot as plt
from pathlib import Path 
import pandas as pd
```

```python
# - Importando os dados diários criados
path_file = Path('../../Dados/Dados-Metereologicos/Diarios/dados_met_diarios_A_criados.csv')
df_bruto = pd.read_csv(path_file)

#Data Medicao agora tipo DateTime
df_bruto['Data Medicao'] = pd.to_datetime(df_bruto['Data Medicao'])

#Tornando a Data Medicao índice
df_bruto.set_index('Data Medicao', inplace=True)

df_bruto.head()
```

```python
#definindo uma função para não precisar ficar repetindo o que for necessário para realizar o plot gráfico
colunas =[ ['Temperatura máxima', 'Velocidade máxima do vento e Rajada Máxima'], 
    ['Precipitação Máxima', 'Radiação máxima']]
def grafico(df, Axes, legenda):
    
    Axes[0,0].plot(df[df.columns[0]], label=legenda)
    Axes[0,1].plot(df[df.columns[1]], label='velocidade do vento máxima ' + legenda)
    Axes[0,1].plot(df[df.columns[2]], label='rajada do vento máxima ' + legenda)
    Axes[1,0].plot(df[df.columns[3]], label=legenda)
    Axes[1,1].plot(df[df.columns[4]], label=legenda)
    
    for i in range(2):
        for j in range(2):
            Axes[i,j].set_title(colunas[i][j])
            Axes[i,j].grid(True)
            
            if legenda != "":
                Axes[i,j].legend()
            else:
                Axes[0,1].legend()
```

```python
#plot do gráfico sem interpolação
fig1, axes1 = plt.subplots(ncols=2, nrows=2, constrained_layout=True, figsize=(20,7))

fig1.suptitle("Sem Interpolação", size=15)

grafico(df_bruto, axes1, "")
```

```python
#fazendo plot da interpolação utilizando linear method
df_linear = df_bruto.interpolate(method= "linear", axis=0)

fig2, axes2 = plt.subplots(ncols=2, nrows=2, constrained_layout=True, figsize=(20,7))
fig2.suptitle("Interpolação linear", size=15)

grafico(df_linear, axes2, "")
```

```python
#fazendo plot da interpolação utilizando polynomial method

df_poly_ordem2 = df_bruto.interpolate(method= "polynomial", order=2, axis=0)
df_poly_ordem3 = df_bruto.interpolate(method= "polynomial", order=3, axis=0)

fig3, axes3 = plt.subplots(ncols=2, nrows=2, constrained_layout=True, figsize=(20,7))
fig3.suptitle("Interpolação Polinomial", size=15)

grafico(df_poly_ordem2, axes3, "ordem 2")
grafico(df_poly_ordem3, axes3, "ordem 3")
```

```python
df_spline =  df_bruto.interpolate(method= "spline", order=2, axis=0)

fig4, axes4 = plt.subplots(ncols=2, nrows=2, constrained_layout=True, figsize=(20,7))
fig4.suptitle("Interpolação por Spline", size=15)

grafico(df_spline, axes4, '')
```

```python
df_pad =  df_bruto.interpolate(method= "pad", axis=0)

fig5, axes5 = plt.subplots(ncols=2, nrows=2, constrained_layout=True, figsize=(20,7))
fig5.suptitle("Interpolação Pad", size=15)

grafico(df_pad, axes5, "")
```

```python
df_derivatives = df_bruto.interpolate(method="from_derivatives", axis=0)


fig6, axes6 = plt.subplots(ncols=2, nrows=2, constrained_layout=True, figsize=(20,7))
fig6.suptitle("Interpolação por derivada", size=15)

grafico(df_derivatives, axes6, "")
```
